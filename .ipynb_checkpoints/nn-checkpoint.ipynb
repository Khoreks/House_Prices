```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv('./data/xdata.csv', index_col='Unnamed: 0')
y = pd.read_csv('./data/ydata.csv', index_col='Id')
test = pd.read_csv('./data/xtest.csv', index_col='Unnamed: 0')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
def print_train(history):
    hist = history.history

    fig = plt.figure(figsize=(12, 5))
    ax = fig.add_subplot(1, 2, 1)
    ax.plot(hist['loss'][100:], lw=3)
    ax.set_title('Training loss', size=15)
    ax.set_xlabel('Epoch', size=15)
    ax.tick_params(axis='both', which='major', labelsize=15)

    ax = fig.add_subplot(1, 2, 2)
    ax.plot(hist['MAE'][100:], lw=3)
    ax.set_title('Training MAE', size=15)
    ax.set_xlabel('Epoch', size=15)
    ax.tick_params(axis='both', which='major', labelsize=15)
    plt.tight_layout()

    plt.show()
    
def print_val(history):
    hist = history.history

    fig = plt.figure(figsize=(12, 5))
    ax = fig.add_subplot(1, 2, 1)
    ax.plot(hist['val_loss'][100:], lw=3)
    ax.set_title('val loss', size=15)
    ax.set_xlabel('Epoch', size=15)
    ax.tick_params(axis='both', which='major', labelsize=15)

    ax = fig.add_subplot(1, 2, 2)
    ax.plot(hist['val_MAE'][100:], lw=3)
    ax.set_title('val MAE', size=15)
    ax.set_xlabel('Epoch', size=15)
    ax.tick_params(axis='both', which='major', labelsize=15)
    plt.tight_layout()

    plt.show()
```

```python
model = Sequential()
model.add(Dense(6000, input_dim=163, kernel_initializer='normal', activation='sigmoid'))
tf.keras.layers.Dropout(0.9)
model.add(Dense(14000, kernel_initializer='normal', activation='softmax'))
tf.keras.layers.Dropout(0.9)
model.add(Dense(3500, kernel_initializer='normal', activation='relu'))
tf.keras.layers.Dropout(0.5)
model.add(Dense(2500, kernel_initializer='normal', activation='relu'))
tf.keras.layers.Dropout(0.5)
model.add(Dense(1350, kernel_initializer='normal', activation='relu'))
tf.keras.layers.Dropout(0.5)
model.add(Dense(800, kernel_initializer='normal', activation='relu'))
tf.keras.layers.Dropout(0.5)
model.add(Dense(380, kernel_initializer='normal', activation='relu'))
tf.keras.layers.Dropout(0.5)
model.add(Dense(250, kernel_initializer='normal', activation='relu'))
tf.keras.layers.Dropout(0.5)
model.add(Dense(180, kernel_initializer='normal', activation='relu'))
model.add(Dense(45, kernel_initializer='normal', activation='relu'))
model.add(Dense(9, kernel_initializer='normal', activation='relu'))
model.add(Dense(4, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal', activation='linear'))

# Compile model
model.compile(loss='mean_absolute_error', optimizer='adam', metrics='MSE')

optimizer = tf.keras.optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
model.compile(optimizer=optimizer, loss='MSE', metrics='MAE')
```

```python
history = model.fit(X, y, epochs=1000, verbose=1,validation_split=0.1)# validation_data=(X_test, y_test))
```
