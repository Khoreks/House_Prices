# Обработка пропусков
Я буду отталкиваться от возможного отсутствия некоторых позиций в домах   
И для сохранения информации присвою им "None"   
Вытекающие числовые характеристики из отсутствующих признаков будут обнулены соответсвенно

```python
# PoolQC: Pool quality
data["PoolQC"] = data["PoolQC"].fillna("None")
# MiscFeature: Miscellaneous feature not covered in other categories
data["MiscFeature"] = data["MiscFeature"].fillna("None")
# Alley: Type of alley access to property
data["Alley"] = data["Alley"].fillna("None")
# Fence: Fence quality
data["Fence"] = data["Fence"].fillna("None")
# FireplaceQu: Fireplace quality
data["FireplaceQu"] = data["FireplaceQu"].fillna("None")

# Для пропусков в признаках гаража предположим его отсутствие
# Качества отделки, состояние гаража и т.д. будет отсутсвовать
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    data[col] = data[col].fillna('None')
# Год построрйки, площадь и машины в гараже занулим
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    data[col] = data[col].fillna(0)
    
# тоже самое для подвала
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    data[col] = data[col].fillna('None')

for col in ('BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath'):
    data[col] = data[col].fillna(0)
#
data["MasVnrType"] = data["MasVnrType"].fillna("None")
data["MasVnrArea"] = data["MasVnrArea"].fillna(0)

# 
data['MSZoning'] = data['MSZoning'].fillna(data['MSZoning'].mode()[0])


#Group by neighborhood and fill in missing value by the median LotFrontage of all the neighborhood
#LotFrontage: Linear feet of street connected to property
print(f'LotFrontage median: {data["LotFrontage"].median()}')
print(f'LotFrontage mean: {data["LotFrontage"].mean()}')

data["LotFrontage"] = data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))




data = data.drop(['Utilities'], axis=1)

data["Functional"] = data["Functional"].fillna("Basic")
# Следующие имеют по 1 пропуску, поэтому просто заменим модой
data['Electrical'] = data['Electrical'].fillna(data['Electrical'].mode()[0])
data['KitchenQual'] = data['KitchenQual'].fillna(data['KitchenQual'].mode()[0])
data['Exterior1st'] = data['Exterior1st'].fillna(data['Exterior1st'].mode()[0])
data['Exterior2nd'] = data['Exterior2nd'].fillna(data['Exterior2nd'].mode()[0])
data['SaleType'] = data['SaleType'].fillna(data['SaleType'].mode()[0])
```

```python
## Проверка пропусков
miss = (data.isnull().sum())
miss = miss.drop(miss[miss == 0].index).sort_values(ascending=False)#[:30]

miss_percent = np.round(((miss / len(data)) * 100),2)

missing_data = pd.DataFrame({'Missing' : miss, '%' : miss_percent})
missing_data
```

```python
# 4 признака имеют категориальный характер
data['MSSubClass'] = data['MSSubClass'].apply(str)
data['OverallCond'] = data['OverallCond'].astype(str)
data['YrSold'] = data['YrSold'].astype(str)
data['MoSold'] = data['MoSold'].astype(str)
```

# Data Encoding 

```python
data_categorical_subset = data.select_dtypes(exclude = [np.number])
data_numerical_subset = data.select_dtypes(include = [np.number])
```

```python
data_categorical_subset = pd.get_dummies(data_categorical_subset)
data_categorical_subset
```

```python
log_data_numerical_subset = np.log(data_numerical_subset)
log_data_numerical_subset = log_data_numerical_subset.replace(to_replace = float('-inf'), value = 0)
log_data_numerical_subset
```

```python
data = pd.concat((log_data_numerical_subset, data_categorical_subset), axis=1)
data
```

```python
train = data.iloc[:len(train)]
test = data.iloc[len(train):]
```

# Modelling

```python
from sklearn.linear_model import ElasticNet, Lasso,  BayesianRidge, LassoLarsIC
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import lightgbm as lgb
```

```python
def rmsle_cv(model):
    kf = KFold(5, shuffle=True, random_state=42).get_n_splits(train)
    rmse= np.sqrt(-cross_val_score(model, train, y, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

```python
# create models
lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.0005))

ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9))

KRR = KernelRidge(alpha=0.6, kernel='polynomial', degree=2, coef0=2.5)

GBoost = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber')

model_xgb = xgb.XGBRegressor(colsample_bytree=0.46, gamma=0.047, 
                             learning_rate=0.05, max_depth=3)

model_lgb = lgb.LGBMRegressor(objective='regression',num_leaves=5,
                              learning_rate=0.05, n_estimators=720,
                              max_bin = 55)
```

```python
# Models score
score = rmsle_cv(lasso)
print("\nLasso score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

score = rmsle_cv(ENet)
print("ElasticNet score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

score = rmsle_cv(KRR)
print("Kernel Ridge score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

score = rmsle_cv(GBoost)
print("Gradient Boosting score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

score = rmsle_cv(model_xgb)
print("Xgboost score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

score = rmsle_cv(model_lgb)
print("LGBM score: {:.4f} ({:.4f})\n" .format(score.mean(), score.std()))
```
