```python
X = pd.read_csv('./data/xdata_prep.csv', index_col='Unnamed: 0')
y = pd.read_csv('./data/ydata_prep.csv', index_col='Unnamed: 0')
test = pd.read_csv('./data/xtest_prep.csv', index_col='Unnamed: 0')
```

```python
X.describe()
```

```python
from sklearn.preprocessing import Normalizer

scaler = Normalizer().fit(X)
X = scaler.transform(X)
test = scaler.transform(test)
```

```python
model = Sequential()
model.add(Dense(4000, input_dim=163, kernel_initializer='normal', activation='relu'))
model.add(Dense(800, kernel_initializer='normal', activation='relu'))
model.add(Dense(150, kernel_initializer='normal', activation='relu'))
#model.add(Dense(100, kernel_initializer='normal', activation='relu'))
model.add(Dense(25, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
model.compile(optimizer='SGD', loss='MSE', metrics='MSE')

num_epochs = 100
training_size = 1094
batch_size = 10
steps_per_epoch = np.ceil(training_size/batch_size)

X_train = tf.data.Dataset.from_tensor_slices(X_train)
y_train = tf.data.Dataset.from_tensor_slices(y_train)
ds_train = tf.data.Dataset.zip((X_train, y_train))

ds_train = ds_train.shuffle(buffer_size=training_size)
ds_train = ds_train.repeat()
ds_train = ds_train.batch(batch_size=batch_size)
ds_train = ds_train.prefetch(buffer_size=1000)

history = model.fit(ds_train, epochs=num_epochs,
                        steps_per_epoch=steps_per_epoch,
                        verbose=1)
```

```python
def larger_model():
# create model
    model = Sequential()
    model.add(Dense(4000, input_dim=163, kernel_initializer='normal', activation='relu'))
    model.add(Dense(800, kernel_initializer='normal', activation='relu'))
    model.add(Dense(150, kernel_initializer='normal', activation='relu'))
    #model.add(Dense(100, kernel_initializer='normal', activation='relu'))
    model.add(Dense(25, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))

    # Compile model
    model.compile(loss='mean_absolute_error', optimizer='adam', metrics='MSE')
    return model
```

```python
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=larger_model, epochs=10000, batch_size=40, verbose=1)))
my = Pipeline(estimators)
```
